# 축 

```python
import matplotlib.pylab as plt

#한글 폰트 설정
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x, y)
plt.title('꺽은선 그래프')
plt.show()
```

```python
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('x축', loc='right') #left, center, right
plt.ylabel('y축', loc='top') #top, center, bottom
plt.show()
```

```python
plt.plot(x, y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```

# 범례(legend)

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend()
plt.show()
```

```python
plt.plot(x, y, label='무슨 데이터')
#plt.legend(loc='upper left')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend(loc=(0.5, 0.5)) # x 축, y축 0~1
plt.show()
```

## 스타일

###  마커

```python
plt.plot(x, y, linewidth=3)
```

```python
#plt.plot(x, y, marker='o')
plt.plot(x, y, marker='o', linestyle='None')
```

```python
#plt.plot(x, y, marker='v', markersize=10)
plt.plot(x, y, marker='x', markersize=10)
```

### 색깔

### 포맷
마커ㅡ 선스타일 등을 한번에 설정

```python
plt.plot(x, y, 'ro--') #color = r, maker = o, linestyle= -- /
```

```python
plt.plot(x, y, 'bv:')
```

```python
#plt.plot(x, y, 'go') #linestyle이 없으면 선이 나타나지 않음
plt.plot(x, y, color='g', marker='o', linestyle='None') # 위와 같은말
```

### 축약어
makerfacecolor => mfc      마커의 색상   
makeredgecolor => mec     마커의 테두리 색상  
markersize => ms    
linestyle => ls

```python
plt.plot(x, y, marker='o', mfc='r', ms = 3, ls='--', mec='r')
```
