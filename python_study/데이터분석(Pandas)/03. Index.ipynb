# Index
데이터에 접근할 수 있는 주소 값

```python
import pandas as pd
```

```python
data = {
    '이름' : ['김자바', '이자바', '박자바', '최자바', '정자바'],
    '학급' : ['자바반', '자바반', '자바반', '캐드반', '캐드반'],
    '국어점수' : [90, 80, 70, 100, 90],
    '영어점수' : [70, 80, 80, 70, 90],
    '평균' : [80.0, 80.0, 75.0, 85.0, 90.0],
    '취미' : ['독서', '', '', '요가', '축구']
}
```

```python
df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번'])
```

```python
df
```

```python
df.index
```

###  Index 이름 설정

```python
df.index.name = '학번'
df
```

### Index 초기화

```python
# index는 지워지지만 index가 컬럼으로 지정되어버림
df.reset_index()
```

```python
df.reset_index(drop=True)
```

```python
df.reset_index(drop=True, inplace=True)
df
```

### Index 설정
지정한 COLUMN으로 INDEX 지정

```python
#df.set_index('이름')
df.set_index('이름', inplace=True)
```

```python
df
```

### Index를 기준으로 정렬

```python
#df.sort_index() #index를 기준으로 오름차순 정렬
df.sort_index(ascending=False)
```
