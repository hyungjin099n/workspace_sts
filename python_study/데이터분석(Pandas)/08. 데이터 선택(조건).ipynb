# 08. 데이터 선택(조건)
조건에 따른 데이터 선택

```python
import pandas as pd
```

```python
df = pd.read_excel('student.xlsx', index_col='학번')
df
```

```python
df['평균'] >= 70
```

```python
myFilter = (df['평균'] >= 80)
df[myFilter]
df[~myFilter] #부정
```

```python
df[df['평균'] >= 80]
```

```python
df.loc[df['평균'] >= 80, '이름']
df.loc[df['평균'] >= 80, ['이름', '평균']]
```

```python
df.loc[(df['평균'] >= 80) & (df['학교'] == '자바학교')]
```

```python
con1 = df['평균'] >= 80
con2 = df['학교'] == '자바학교'
df.loc[con1 & con2]
```

```python
df.loc[(df['수학점수'] < 80) | (df['수학점수'] > 90)]
```

### str 함수

```python
myFilter = df['이름'].str.startswith('김') #송씨 성을 가진 사람
df[myFilter]
```

```python
myFilter = df['이름'].str.contains('바') #바자를 가진 사람
df[myFilter]
df[~myFilter]
```

```python
hobby = ['운동', '독서']
myFilter = df['취미 '].isin(hobby)
df[myFilter]
```

```python
hobby = ['운동', '독서']
myFilter = df['취미 '].str.lower().isin(hobby) #대소문자 구분 없애줌
df[myFilter]
```

```python
myFilter = df['취미 '].str.contains('독서', na=True) #Nan 데이터는 전부 True
myFilter
df[myFilter]

#구글에서  pandas string 검색 시 사용가능한 함수 정보 나옴
```
