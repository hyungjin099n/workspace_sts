## 데이터 채우기 fillna

```python
df.fillna('') #NaN 데이터를 빈 캄으로 채움
```

```python
df.fillna('없음') #NaN 데이터를 없음으로 채움
```

```python
import numpy as np #수학과 관련된 여러 기능을 제공하는 라이브러리
df['학급'] = np.nan #학급 데이터 전체를 NaN으로 채움
df
```

```python
df.fillna('모름', inplace=True)
df
```

##  데이터 제외하기(dropna)

```python
df = pd.read_excel('student.xlsx', index_col='학번')
df
```

```python
df.dropna() # NaN 데이터(행 전체)를 삭제
df
```

```python
df.dropna(inplace=True) # NaN 데이터(행 전체)를 삭제
df
```

```python
df = pd.read_excel('student.xlsx', index_col='학번')
df
```

```python
# axis = 'index' : NaN 데이터를 기준으로 행을 삭제 (default)
# axis = 'columns' : NaN 데이터를 기준으로 열을 삭제
# how = 'any' : NaN이 하나라도 있으면.. (default)
# how = 'all' : 모두 NaN이면...
df.dropna(axis='index', how='any') #NaN이 하나라도 있는 ROW 삭제
```

```python
df.dropna(axis='columns') #NaN이 하나라도 있는 column 삭제
```

```python
df = pd.read_excel('student.xlsx', index_col='학번')
df
```
