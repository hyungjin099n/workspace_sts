# 데이터 정렬

```python
import pandas as pd
df = pd.read_excel('student.xlsx', index_col='학번')
df
```

```python
df.sort_values('평균')#오름차순
df.sort_values('평균', ascending=False) #내림차순
```

```python
df.sort_values(['국어점수', '영어점수'], ascending=False)
```

```python
#국어점수는 오름차순, 영어점수는 내림차순
df.sort_values(['국어점수', '영어점수'], ascending=[True, False])
```

```python
#df['평균'].sort_values()
df['평균'].sort_values(ascending=False)
```
